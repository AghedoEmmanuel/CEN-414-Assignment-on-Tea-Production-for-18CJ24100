```python
iq_vals = []
def get_iq(sample):
    q3,q1 = np.percentile(sample,[75,25])
    return q3 - q1
for year in range(2010,2019):
    iq_vals.append(get_iq(df[df['Year']==year]['Value']))
    

plt.ylabel("Ratio")
plt.xlabel("Years")
plt.title("Interquatile Range of Tea Production in Africa Over the Years")
plt.plot(range(2010,2019),iq_vals)
```
